## Question 17, 20, 21, 22

```python
import pandas as pd
```

```python
# print question 17, 20, 21, 22
salary_data = pd.read_csv('kaggle_survey_2020_responses.csv')

Q_17_20_21_22 = salary_data.iloc[:, list(range(82,94)) + list(range(107, 110))]
Q_17_20_21_22
```

```python
# check the uniquess of each columns to see whether there exits some weried answer
for col in Q_17_20_21_22:
    print(Q_17_20_21_22[col].unique())
```

## Clean data with one column questions: Use one hot encoding

```python
# select the question with just one column
Q_20_21_22 = Q_17_20_21_22[['q20', 'q21','q22']] 

# check the number of null values
missing_values = Q_20_21_22.isna().sum()/Q_20_21_22.shape[0]
missing_values
```

```python
# replace missing value with the category that occurs most frequently

Q_20_21_22['q20'].fillna(Q_20_21_22['q20'].mode()[0], inplace=True)
Q_20_21_22['q21'].fillna(Q_20_21_22['q21'].mode()[0], inplace=True)
Q_20_21_22['q22'].fillna(Q_20_21_22['q22'].mode()[0], inplace=True)
Q_20_21_22
```

```python
# one hot encoding for just question just having one colomn
cleaned_Q_20_21_22 = pd.get_dummies(Q_20_21_22, columns = ['q20', 'q21','q22'], drop_first = True)
cleaned_Q_20_21_22
```

## Clean data with multiple column questions: Replace all non-NaN entries of a dataframe with 1 and all NaN with 0

```python
# drop the question with just one column and we will do data clean with other quesitons excpet them
Q_17 = Q_17_20_21_22.drop(['q20', 'q21','q22'], axis=1)
Q_17
```

```python
# replace all non-NaN entries of a dataframe with 1 and all NaN with 0
newdf = Q_17.notnull().astype('int')
#newdf.drop('q17_other')
newdf

#newdf.drop('q17_other')
cleaned_Q_17 = newdf.drop(['q17_other'], axis=1)
cleaned_Q_17
```

## Combine these two parts into one dataframe

```python
Combined_Cleaned_Q_17_20_21_22 = pd.concat([cleaned_Q_17, cleaned_Q_20_21_22], axis=1, join='inner')
Combined_Cleaned_Q_17_20_21_22
```

```python
Combined_Cleaned_Q_17_20_21_22.to_csv('Lin_CleanedData_Q17_20_21_22.csv', index=False)
```
